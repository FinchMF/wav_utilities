## Navigating Alternative Musical Systems

```python
from freq_table import *
from wav_table import *
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
# starting with the western semi tone systems, lets generate an entire system as a base system
interval_systems.get('semi_tones')
```

```python
# instantiate a C -1 class
C_neg_1 = Hz(pitch_to_frequency.get('C')[-1])
# make a semi_tone system with 128 frequencies 
semitone_sys_Cn1 = C_neg_1.make_system(system_type=interval_systems.get('semi_tones'), 
                                       system_size=range(0,128))
semitone_sys_Cn1                    
```

This is a majority of the western musical system. I chose to build out the system to 128 frequencies just incase we want to convert these frequencies in to MIDI notes later. 

```python
# using the closest pitch function, we are able to see the pitch names in scientific notation of each frequency
for freg, hz in semitone_sys_Cn1.items():
    print(closest_pitch(hz))
```

```python
# lets zoom in on the octave we have already been working with and orientate it to start at A4

# instantiate an A4  class
A_4 = Hz(pitch_to_frequency.get('A')[4])
# make a semi_tone system with 12 frequencies 
semitone_sys_A_4 = A_4.make_system(system_type=interval_systems.get('semi_tones'), 
                    system_size=range(0,12))
semitone_sys_A_4 
```

```python
# lets make a base analysis of the frequencies present in the octave this semitone system
semi_tone_sys_analysis = develop_offset_dict(semitone_sys_A_4)
```

We can see that each frequency directly corresponds to the frequency of a pitch. There is zero cent offset. 

```python
semi_tone_sys_analysis
```

```python
# visualizing the cent differences
y = list(semi_tone_sys_analysis.values())
x = list(semi_tone_sys_analysis.keys())
plt.figure(figsize=(20,10))
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.hlines(y=0, 
           xmin=0, 
           xmax=len(x), 
           colors='r', 
           linestyles='--', 
           lw=2)
ax.plot(x, y)
plt.hlines(y=0, 
           xmin=0, 
           xmax=len(x), 
           colors='r', 
           linestyles='--', 
           lw=2)

plt.show()
```

This graph visualizes that each frequency in the dictionary 'semi_tone_sys_analysis' is centered directly corresponding with the pitch's frequency. There is no cent offset between the generated given frequency and the frequency of the pitch. We can use this visualization as a base to show how the other systems differ.

## Alternative System 1 (tri_tone system)

```python
# now lets look at a system that divides the octave into 18 intervals rather than 12

# make a tri_tone system with 18 frequencies 
tritone_sys_A_4 = A_4.make_system(system_type=interval_systems.get('tri_tones'), 
                    system_size=range(0,19))
tritone_sys_A_4 
```

```python
print('Here we can see the octave starting at A4 and each frequency corresponding to each of the resultant 18 pitches')
for factor in range(0,19):
    print(A_4.hz * (interval_systems.get('tri_tones')**factor))
print('\n')
print('Here we can see the factors of each interval that divide the octave')
for factor in range(0,19):
    print(interval_systems.get('tri_tones')**factor)
```

```python
# now lets analyze each pitch and distance of each division in cents relative to the western base system 12 
tri_tone_sys_analysis = develop_offset_dict(tritone_sys_A_4)
```

```python
tri_tone_sys_analysis
```

```python
# visualizing the cent differences
y = list(tri_tone_sys_analysis.values())
x = list(tri_tone_sys_analysis.keys())
plt.figure(figsize=(20,10))
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.hlines(y=0, 
           xmin=0, 
           xmax=len(x), 
           colors='r', 
           linestyles='--', 
           lw=2)
ax.plot(x, y)
plt.hlines(y=0, 
           xmin=0, 
           xmax=len(x), 
           colors='r', 
           linestyles='--', 
           lw=2)

plt.show()
```

In this system we can see that 6 notes from the western base 12 system are present and that every other pitch is roughly 33 cents sharp or flat from one of those 6 notes. 

The six notes per octave that are common between the base 12 system and the base 18 system area:

[0, 2, 4, 6, 8, 10].

In diatonic language:
- The Tonic
- The Major Second
- The Major Third
- The Sharp Fourth
- The Flat Sixth
- The Flat Seventh

## Alternative System 3 (sixth_tone system)

```python
# now lets look at a system that divides the octave into 36 intervals

# make a sixth_tone system with 36 frequencies 
sixthtone_sys_A_4 = A_4.make_system(system_type=interval_systems.get('sixth_tones'), 
                                      system_size=range(0,37))
sixthtone_sys_A_4 
```

```python
print('Here we can see the octave starting at A4 and each frequency for each divison of 36')
for factor in range(0,37):
    print(A_4.hz * (interval_systems.get('sixth_tones')**factor))
print('\n')
print('Here we can see the factors of each interval that divide the octave')
for factor in range(0,37):
    print(interval_systems.get('sixth_tones')**factor)
```

Interesting to note that while this system incoporates all 12 pitches from the western system, it does not fully reach the octave. If rounded, it would - but the natural intervals do not quite make it.

```python
# now lets analyze each pitch and distance of each division in cents relative to the western base system 12 
sixth_tone_sys_analysis = develop_offset_dict(sixthtone_sys_A_4)
```

We can see that there is a realtionship between the tri_tone system and the sixth_tone system. Just as there is a relationship between the semi_tone system and the quarter_tone system. 

The relation being that the former are systems built from divions of 3s, while the latter is built from divions of 2s

Where the tri_tone system uses 6 notes from the western system and divides each interval into thrids, the sixth_tone system take all 12 notes and divides each interval into thirds

```python
sixth_tone_sys_analysis
```

```python
# visualizing the cent differences
y = list(sixth_tone_sys_analysis.values())
x = list(sixth_tone_sys_analysis.keys())
plt.figure(figsize=(20,10))
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.hlines(y=0, 
           xmin=0, 
           xmax=len(x), 
           colors='r', 
           linestyles='--', 
           lw=2)
ax.plot(x, y)
plt.hlines(y=0, 
           xmin=0, 
           xmax=len(x), 
           colors='r', 
           linestyles='--', 
           lw=2)

plt.show()
```

A system within a system. 

The point of this exercise has two main objectives. 

1. To show how to generate useable systems of frequencies and compare them to the system we are most familiar with. These frequencies can then be used to generate music with similar conventions that are used with the western system.

2. To show that that the space between frequencies is infinitely divisible. To show that the frequency spectrum is a contionus space, that can be discretely divided in order to manipulate and become a medium of symbolic expression
