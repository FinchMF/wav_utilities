## Examining Wav Files

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

## Pulling From Signal Modulation Class

```python
rec = 'simple_sin.wav'
rec_2 = 'amplitude_modulated.wav'
```

```python
y, sr = librosa.load(rec, duration=10)
plt.figure()
plt.subplot(3, 1, 1)
librosa.display.waveplot(y, sr=sr)
plt.title('simple_sine')
```

```python
y, sr = librosa.load(rec_2, duration=10)
plt.figure()
plt.subplot(3, 1, 1)
librosa.display.waveplot(y, sr=sr)
plt.title('amplitude_modulated')
```
