## Examining Wav Files

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

## Pulling From Signal Modulation Class

```python
ls
```

```python
sin_rec = 'simple_sin.wav'
mod_sin_rec_ = 'mod_sin.wav'

sq_rec = 'simple_square.wav'
mod_sq_rec = 'mod_sq.wav'

saw_rec = 'simple_sawtooth.wav'
mod_saw_rec = 'mod_saw.wav'
```

```python
y, sr = librosa.load(sin_rec, duration=10)
plt.figure()
plt.subplot(3, 1, 1)
librosa.display.waveplot(y, sr=sr)
plt.title('simple_sine')
```

```python
y, sr = librosa.load(mod_sin_rec_, duration=10)
plt.figure()
plt.subplot(3, 1, 1)
librosa.display.waveplot(y, sr=sr)
plt.title('sin_modulated')
```

```python
y, sr = librosa.load(mod_sq_rec, duration=10)
plt.figure()
plt.subplot(3, 1, 1)
librosa.display.waveplot(y, sr=sr)
plt.title('square_modulated')
```

```python
y, sr = librosa.load(mod_saw_rec, duration=10)
plt.figure()
plt.subplot(3, 1, 1)
librosa.display.waveplot(y, sr=sr)
plt.title('sawtooth_modulated')
```
